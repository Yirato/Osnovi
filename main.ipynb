```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('Top100-2007.csv')
df
```

```python
df.shape
```

```python
K = 5
pd.concat([df.iloc[K:K + 5, :], df.iloc[-(3 * K + 2):, :]])
```

```python
df.dtypes
```

```python
for column in ['Name', 'Country', 'Singles Record (Career)', 'Winning Percentage', 'Career Earnings', 'Link to Wikipedia']:
    df[column] = df[column].map(lambda x: x.strip() if not pd.isna(x) else x)
```

```python
df['Winning Percentage'] = df['Winning Percentage'].map(lambda x: float(x[:-1]) if not pd.isna(x) else x)
df['Career Earnings'] = df['Career Earnings'].map(lambda x: int(x[1:]) if not pd.isna(x) else x)

df
```

```python
df[df.isna().any(axis=1)]
```

```python
df = df.dropna(axis=0)
```

```python
df['Win'] = df['Singles Record (Career)'].map(lambda x: int(x.split('-')[0]))
df['Lose'] = df['Singles Record (Career)'].map(lambda x: int(x.split('-')[1]))
df['Total'] = df['Win'] + df['Lose']
```

```python
df = df.drop(['Singles Record (Career)', 'Link to Wikipedia'], axis=1)
```

```python
df = df[['Rank', 'Name', 'Country', 'Pts', 'Total', 'Win', 'Lose', 'Winning Percentage', 'Career Earnings']]
```

```python
df['Country'].sort_values().unique()
```

```python
df[df['Career Earnings'] == df['Career Earnings'].min()][['Name', 'Pts']]
```

```python
df[df['Win'] == df['Lose']][['Name', 'Country']]
```

```python
df.groupby('Country')['Country'].count()
```

```python
df[['Country', 'Rank']].groupby('Country').mean()
```

```python
df1 = df[['Rank', 'Lose']].sort_values('Rank')
groupes = []
loses = []
for i in range(0, 91, 10):
    groupes.append(f'{i + 1}-{i + 10}')
    loses.append(df1.iloc[i:i+10]['Lose'].sum())
groupes[-1] = '91-98'

plt.bar(groupes, loses)
plt.show()
```

```python
df1 = df[['Country', 'Career Earnings']].groupby('Country').sum()
plt.pie(df1['Career Earnings'], labels=df1.index, textprops={'fontsize': 8})
plt.show()
```

```python
df1 = df[['Country', 'Total', 'Pts']].groupby('Country').mean()
plt.scatter(df1['Total'], df1['Pts'])
for i, country in enumerate(df1.index): 
    plt.annotate(country, (df1['Total'].iloc[i], df1['Pts'].iloc[i]), fontsize=8)

plt.show()
```
